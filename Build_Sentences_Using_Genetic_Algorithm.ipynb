```python
import numpy as np

tujuan = "Aku punya ini"
ukuran = 50
banyak_generasi = 1000000
size_tur = 5
chance_crossover = 0.8
chance_dimutasi = 0.1
len_huruf = len(tujuan)
target_fit = len(tujuan)

def fit(huruf):
    return sum(1 for a, b in zip(huruf, tujuan) if a == b)

def populate(ukuran, len_huruf):
    return [''.join(np.random.choice(list('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ '), size=len_huruf)) for _ in range(ukuran)]

def turnamen(populasi, skor_fit, size_tur):
    parent_terpilih = []
    for _ in range(len(populasi)):
        index = np.random.choice(len(populasi), size=size_tur, replace=False)
        fit = [skor_fit[i] for i in index]
        parent_terpilih.append(populasi[index[np.argmax(fit)]])
    return parent_terpilih

def crossover(parent1, parent2):
    titik_crossover = np.random.randint(1, len(parent1) - 1)
    child1 = parent1[:titik_crossover] + parent2[titik_crossover:]
    child2 = parent2[:titik_crossover] + parent1[titik_crossover:]
    return child1, child2

def mutasi(child, chance_dimutasi):
    child_termutasi = list(child)
    for i in range(len(child_termutasi)):
        if np.random.rand() < chance_dimutasi:
            child_termutasi[i] = np.random.choice(list('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ '))
    return ''.join(child_termutasi)

def algoritma_genetika(ukuran, banyak_generasi, size_tur, chance_crossover, chance_dimutasi, len_huruf, target_fit):
    populasi = populate(ukuran, len_huruf)

    solusi = ""
    fit_final = 0

    for generasi in range(banyak_generasi):
        skor_fit = [fit(huruf) for huruf in populasi]
        solusi_temp = max(populasi, key=fit)
        fit_temp = fit(solusi_temp)

        print(f"Generasi {generasi + 1}\t| Hasil kata: {solusi_temp}\t| Fit: {fit_temp}")

        if fit_temp >= target_fit:
            print(f"\nBerhasil di generasi {generasi + 1}")
            break

        parent = turnamen(populasi, skor_fit, size_tur)
        for i in range(0, len(parent), 2):
            if np.random.rand() < chance_crossover:
                parent1, parent2 = parent[i], parent[i+1]
                child1, child2 = crossover(parent1, parent2)
                parent[i], parent[i+1] = child1, child2

        for i in range(len(parent)):
            parent[i] = mutasi(parent[i], chance_dimutasi)

        populasi = parent
    return solusi

solusi = algoritma_genetika(ukuran, banyak_generasi, size_tur, chance_crossover, chance_dimutasi, len_huruf, target_fit)
```
